## Perform Exploratory Data Analysis (EDA) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
sys.path.append(os.path.abspath('../data'))
```

```python
file_path = '../data/MachineLearningRating_v3.txt'
```

```python
df = pd.read_csv(file_path, sep='|')
```

```python
# Display the first few rows of the DataFrame
print(df.head())
```

## Data Summarization

```python
df.describe()
```

```python
# Display the first Info of dataframe
print(df.info())
```

## Descriptive statistics for TotalPremium, TotalClaims, SumInsured, CalculatedPremiumPerTerm

```python
# Descriptive statistics for numerical columns
numerical_columns = ['TotalPremium', 'TotalClaims', 'SumInsured', 'CalculatedPremiumPerTerm']

# Descriptive statistics
df[numerical_columns].describe()
```

## Checking data type

```python
# Checking the data types
df.dtypes
```

## Converting datatype into date time

```python
# Conerting datatype into date time
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth']) # TransactionMonth to datetime
df['VehicleIntroDate'] = pd.to_datetime(df['VehicleIntroDate'])  # VehicleIntroDate to datetime
#df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
# Checking the data types
df.dtypes
```

## Convert categorical columns into category

```python
# Ensuring categorical columns are converted to 'category' dtype
# In this case I use the the data type of object to category
categorical_columns = ['Citizenship','LegalType','Title','Language','Bank','AccountType','MaritalStatus', 'Gender', 'Country','Province','MainCrestaZone','SubCrestaZone','ItemType','VehicleType','make','Model','bodytype','AlarmImmobiliser','TrackingDevice','CapitalOutstanding','NewVehicle','WrittenOff','Rebuilt','Converted','TermFrequency','ExcessSelected','CoverCategory', 'CoverType', 'CoverGroup', 'Section', 'Product', 'StatutoryClass', 'StatutoryRiskType']
df[categorical_columns] = df[categorical_columns].astype('category')
```

## Data Quality Assessment

```python
# Checking for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
# I have missing value in the data but the questions say that just see the missing value 
# not replace the missing so I did not to on the misssing value
```

## Univariate Analysis

```python
# Plotting histograms for numerical columns
df[numerical_columns].hist(bins=15, figsize=(15, 10))
plt.show()
```

```python
# Bar charts for categorical columns
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.show()
```

## Bivariate or Multivariate Analysis

```python
# First lets do the Correlation matrix between numerical columns
correlation_matrix = df[['TotalPremium', 'TotalClaims', 'SumInsured']].corr()
```

```python
# Heatmap to visualize correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Premiums and Claims')
plt.show()
```

```python
# Scatter plot of TotalPremium vs TotalClaims by Zipcode in this case we don't have zipcode so we use PostalCode
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='PostalCode', data=df, palette='viridis')
plt.title('Total Premium vs Total Claims by PostalCode')
plt.show()
```

## Data Comparsion

```python
# Grouping data by CoverType and PostalCode
grouped_data = df.groupby(['PostalCode', 'CoverType']).agg({
    'TotalPremium': 'mean',
    'SumInsured': 'mean'
}).reset_index()
```

```python
# Visualizing the trends
plt.figure(figsize=(15, 8))
sns.lineplot(x='PostalCode', y='TotalPremium', hue='CoverType', data=grouped_data)
plt.title('Average Premium by CoverType and PostalCode')
plt.show()
```

## Outlier Detection

```python
# Using box plot for numerical variables to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_columns])
plt.title('Boxplot for Numerical Columns')
plt.show() 
```

## Visualization

```python

# Setting up the figure for the two histograms for the TotalPremium and TotalClaims
plt.figure(figsize=(14, 6))

# Histogram for TotalPremium
plt.subplot(1, 2, 1)
sns.histplot(df['TotalPremium'], bins=30, kde=True, color='blue')
plt.title('Histogram of TotalPremium')
plt.xlabel('Total Premium')
plt.ylabel('Frequency')

# Histogram for TotalClaims
plt.subplot(1, 2, 2)
sns.histplot(df['TotalClaims'], bins=30, kde=True, color='green')
plt.title('Histogram of TotalClaims')
plt.xlabel('Total Claims')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Grouping data by PostalCode and CoverType
grouped_data = df.groupby(['PostalCode', 'CoverType']).agg({
    'TotalPremium': 'mean'
}).reset_index()

# Line plot for TotalPremium changes across PostalCode, categorized by CoverType
plt.figure(figsize=(12, 8))
sns.lineplot(x='PostalCode', y='TotalPremium', hue='CoverType', data=grouped_data, marker='o')
plt.title('Average Total Premium by CoverType Across Postal Codes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Clustered bar plot for CoverType across Postal Codes
plt.figure(figsize=(12, 8))
sns.barplot(x='PostalCode', y='TotalPremium', hue='CoverType', data=grouped_data)
plt.title('CoverType Trends by PostalCode')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
